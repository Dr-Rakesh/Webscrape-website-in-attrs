# Webscraping top 50 youtubers
![image](https://i.imgur.com/69cJQvG.png)

```python
#Request to pass the page with user agent which is going to access the website as a browser not as a automated scraper.
request = urllib2.Request(url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.124 Safari/537.36 Edg/102.0.1245.44'})
page = urllib2.urlopen(request)
#For beautiful soup 'html.parser' is the best parser out of many parsers avilable
soup = BeautifulSoup(page, 'html.parser')
```

# **Narrow down the search**
- We are trying to narrow down our scraping. i.e., instead of scraping data from the entire website to get the specific data. We instead are going to select the specific table under which our data is to be scraped.

```python
#Finding all tags of div
all_link_tags = soup.find_all('div')
```

```python
all_link_tags
```

#  Finding page title tag

```python
#Finding title tag
title_tag = soup.title
title_tag 
```

```python
title_tag.text
```

```python
# We use find, as we are secific about our search and we are using attrs since the website uses inline styles.
# We are going to eliminate some unnecessary data, so we are using "find_all('div', recursive=False)[4:]"
channels_tag = soup.find('div', attrs={'style': 'float: right; width: 900px;'}).find_all('div', recursive=False)[4:]
```

```python
channels_tag
```

```python

#Since we want to save the scraped data we are opening a CSV file and wrting the data to be stored. 
file = open('topyoutubers.csv', 'w')
writer = csv.writer(file)

#We are labeling the columns under which we want all our data to be store
writer.writerow(['Username', 'Uploads', 'Views'])
title = title_tag.text
print (title)
#Starting the for loop. Which is going to scrap the individual data 
for channel in channels_tag:
    username = channel.find('div', attrs={'style': 'float: left; width: 350px; line-height: 25px;'}).a.text.strip().replace('b','')
    uploads = channel.find('div', attrs={'style': 'float: left; width: 80px;'}).span.text.strip().replace('b','')
    views = channel.find_all('div', attrs={'style': 'float: left; width: 150px;'})[1].span.text.strip().replace('b','')

#We print each of these scraped data under the given label and also going to write down the data in a csv file 
    print (username + ' ' + uploads + ' ' + views)
    writer.writerow([username.encode('utf-8'), uploads.encode('utf-8'), views.encode('utf-8')])
 
file.close()
```

```python
#Display of recorded data in data Frame
topyoutubers_data = pd.read_csv('topyoutubers.csv')
topyoutubers_data
```

# Webscraping top sorted by followers twitter accounts
![image](https://i.imgur.com/eIxS8nn.png)

```python
#Write URL separately so that it is easier to edit or change in case necessary.
url1= 'https://socialblade.com/twitter/top/100'
```

```python
#Request to pass the page with user agent which is going to access the website as a browser not as a automated scraper.
request = urllib2.Request(url1, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.124 Safari/537.36 Edg/102.0.1245.44'})
page = urllib2.urlopen(request)
#For beautiful soup 'html.parser' is the best parser out of many parsers avilable
soup1 = BeautifulSoup(page, 'html.parser')
```

# **Narrow down the search**
- We are trying to narrow down our scraping. i.e., instead of scraping data from the entire website to get the specific data. We instead are going to select the specific table under which our data is to be scraped.

```python
#Finding all tags of div
all_link_tags = soup1.find_all('div')
```

# Finding page title tag

```python
#Finding title tag
title_tag = soup1.title
title_tag 
```

```python
# We use find, as we are secific about our search and we are using attrs since the website uses inline styles.
twitter_tag = soup1.find('div', attrs={'style': 'float: right; width: 900px;'}).find_all('div', recursive=False)[4:]
```

```python
twitter_tag
```

```python
#Since we want to save the scraped data we are opening a CSV file and wrting the data to be stored. 
file = open('toptwitter.csv', 'w')
writer = csv.writer(file)

#We are labeling the columns under which we want all our data to be store
writer.writerow(['Username', 'Uploads', 'Followers', 'Following'])
title = title_tag.text
print (title)
#Starting the for loop. Which is going to scrap the individual data 
for twitter in twitter_tag:
    username = twitter.find('div', attrs={'style': 'float: left; width: 200px; line-height: 25px;'}).a.text.strip()
    uploads = twitter.find('div', attrs={'style': 'float: left; width: 80px;'}).span.text.strip()
    followers = twitter.find_all('div', attrs={'style': 'float: left; width: 100px;'})[0].span.text.strip()
    following = twitter.find_all('div', attrs={'style': 'float: left; width: 100px;'})[1].span.text.strip()
    
#We print each of these scraped data under the given label and also going to write down the data in a csv file 
    print (username + ' ' + uploads + ' ' + followers  + ' ' + following)
    writer.writerow([username.encode('utf-8'), uploads.encode('utf-8'), followers.encode('utf-8'), following.encode('utf-8')])
 
file.close()
```

```python
#Display of all recorded data
toptwitter_data = pd.read_csv('toptwitter.csv')
toptwitter_data
```
